```python
import numpy as np
import math
from sklearn.linear_model import LinearRegression



# def predict_using_sklearn():
#     r = LinearRegression()
#     r.fit(x,y)
#     return r.coef_, r.intercept_


def gradient_descent(x,y):
    
    m = b = 0
    n = len(x)
    iterations = 80
    learning_rate = 0.0002
    cost_previous = 0

    for i in range(iterations):

        y_predicted = m*x + b 
        
        cost = (1/n)*sum([val*2 for val in (y-y_predicted)])

        md = -(2/n)*sum(x*(y-y_predicted))
        bd = -(2/n)*sum(y-y_predicted)

        m = m - learning_rate * md
        b = b - learning_rate * bd
        
        if math.isclose(cost, cost_previous, rel_tol=1e-20):
        
            break 
        
        cost_previous = cost
            
        print('m {}, b {}, cost {}, iterations{}'.format(m,b,cost,i))
        
    return m,b
        
x = np.array([92,56,88,70,80,49,65,35,66,67])
y = np.array([98,68,81,80,83,52,66,30,68,73])

# x = x.reshape(-1,1)
gradient_descent(x,y)



# # gradient_descent(x,y)
# # print("Using gradient descent function: Coef {} Intercept {}".format(m_curr, b_curr))

# m_sklearn, b_sklearn = predict_using_sklearn()
# print("Using sklearn: Coef {} Intercept {}".format(m_sklearn,b_sklearn))
```
